```python
import pandas as pd
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
df2 = pd.read_csv(io.BytesIO(uploaded['data2.csv']))
```

```python
df2.head()
```

```python
drop = ['State/County FIPS',]
df3 = df2.drop(columns=drop)
print(df3.head())
```

```python
df4 = df3.groupby(['State Name']).sum()
df4.reset_index(inplace=True)
df4.head()

```

```python
df4.head()
```

```python
df5=df4.sort_values('Total', ascending=False)
df5.head()
```

```python
df5.head(10)
```

```python
df6=df5.head(10)
df6
```

```python
df6.to_csv('data5.csv', index=True)
```
